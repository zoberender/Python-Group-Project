```python
# Importing Pandas
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Importing Data
Batting = pd.read_csv("https://raw.githubusercontent.com/zoberender/Python-Group-Project/master/Batting.csv")
Batting
```

```python
# Matplotlib 1
# In any given year, the max number of HR's from any team.
Batting2 = Batting[(Batting['yearID'] > 1980)]
Batting2 = Batting2[['yearID', 'teamID', 'HR']]

Batting2 = Batting2.groupby('yearID').max().reset_index()

labels = Batting2['yearID']
HRs = Batting2['HR']
width = .35

fig, ax = plt.subplots()

ax.bar(labels,HRs, width, label='')
ax.set_ylabel('HRs')
ax.set_title('Max HR per Year')
ax.legend()

plt.show()
```

```python
# Matplotlib 2
# Top HR's by team
Batting3 = Batting[(Batting['yearID'] >= 2015)]
Batting3 = Batting3[['yearID', 'teamID', 'HR']]
Batting3 = Batting3.groupby(['yearID','teamID']).max().reset_index()

teams = Batting3['teamID'].unique()

years = [2015, 2016, 2017, 2018, 2019, 2020]
import numpy.random as rnd
npTemps = np.zeros([len(years), len(teams)])

Hrs = Batting3[['HR']] #getting only the Homeruns
Hrs = np.array(Hrs)  #converting to array
npTemps = np.reshape(Hrs, (6, 30)) #reshaping array

fig, ax = plt.subplots()
im = ax.imshow(npTemps)

# We want to show all ticks...
ax.set_xticks(np.arange(len(teams)))
ax.set_yticks(np.arange(len(years)))
# ... and label them with the respective list entries
ax.set_xticklabels(teams)
ax.set_yticklabels(years)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(teams)):
    for j in range(len(years)):
        text = ax.text(i, j, npTemps[j, i],
                       ha="center", va="center", color="w")

ax.set_title("Max HR per Year")
fig.tight_layout()
plt.show()
```

```python
#seaborn 1
# Total HRs per team going back to 1900.
import seaborn as sns
sns.set_theme(style="white")

Batting4 = Batting[(Batting['yearID'] >= 1900)]
Batting4 = Batting4[['yearID', 'teamID', 'HR']]
Batting4 = Batting4.groupby(['yearID','teamID']).sum().reset_index()

sns.relplot(x="yearID", y="HR", hue="teamID", size="HR",
            sizes=(100, 400), alpha=.5, palette="muted",
            height=6, data=Batting4)
```

```python
#seaborn 2
# Max HR's by player & team 2015-2020
sns.set_theme(style="darkgrid")

# Plot the responses for different events and regions
sns.lineplot(x="yearID", y="HR",
             hue="teamID",
             data=Batting3)
```

```python
#plotly 1
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import numpy as np
Batting = pd.read_csv("https://raw.githubusercontent.com/zoberender/Python-Group-Project/master/Batting.csv")
Batting5 = Batting[(Batting['yearID'] >= 1900)]
Batting5 = Batting5[['yearID', 'teamID', 'HR']]
Batting5 = Batting5.groupby(['yearID']).sum().reset_index()

# Create figure
fig = go.Figure()

fig.add_trace(
    go.Scatter(x=list(Batting5['yearID'].unique()), y=list(Batting5['HR'])))

# Set title
fig.update_layout(
    title_text="Total HRs per Year"
)

# Add range slider
fig.update_layout(
    xaxis=dict(
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)

fig.show() #sum of all the HRs per year.
```

```python
#plotly 2
Batting6 = Batting[(Batting['yearID'] >= 1980)]
Batting6 = Batting6[['yearID', 'playerID', 'teamID', 'SO']]
Batting6 = Batting6.groupby(['yearID', 'teamID' , 'playerID']).sum().reset_index()
Batting6 = Batting6.groupby(['yearID']).max().reset_index()
print(Batting6)

#Can use either or, both provide good results
#fig2 = px.bar(Batting6, y=list(Batting6['SO']), x=list(Batting6['yearID'].unique()), text='SO', color='playerID')
fig2 = px.bar(Batting6, y=list(Batting6['SO']), x=list(Batting6['yearID'].unique()), text='SO', color='teamID')

fig2.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig2.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
fig2.show() #Most strikeouts per year
```
